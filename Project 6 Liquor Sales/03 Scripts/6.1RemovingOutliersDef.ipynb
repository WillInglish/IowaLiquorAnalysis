```python
def remove_column_outliers(df, col):
    
    # IQR
    # Calculate the upper and lower limits
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
 
    # Printing bounds
    print(col, '\nLower bound is: ', lower)
    print('Upper bound is: ', upper)
    print('Median is: ', df[col].median())
    print(df.shape, '\n')
    # Create arrays of Boolean values indicating the outlier rows
    upper_array = np.where(df[col]>=upper)[0]
    lower_array = np.where(df[col]<=lower)[0]
 
    # Removing the outliers
    df.drop(index=upper_array, inplace=True)
    df.drop(index=lower_array, inplace=True)
    df.reset_index(drop=True, inplace=True)
    
    #displaying boxplot of column without outliers
    # df.boxplot(column=[col])
```

```python
for x in df_sales[['bottles_sold', 'sale_dollars', 'package_size', 'volume_sold_liters']]:
    remove_column_outliers(df_sales,x)
```
